```python
import pandas as pd
```

```python
df = pd.DataFrame(
    {
        "Enseignement": [
            "Traitement de l'image",
            "Traitement de l'image",
            "Traitement du signal",
            "Système de calcul formel",
            "Imagerie médicale",
            "Imagerie médicale",
        ],
        "Niveau": [
            "Master 1",
            "Master 1",
            "Master 1",
            "Licence 3",
            "Master 1",
            "Master 1",
        ],
        "Volume horaire (heures)": [
            12, 60, 42, 36, 20, 32
        ],
        "Type d'enseignement": [
            "TD", "TP", "TP", "TP", "TD", "TP"
        ]
    }
)
```

```python
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_context("poster")
```

```python
fig, axs = plt.subplots(ncols=2, figsize=(20, 5))

g = sns.barplot(
    data=df,
    x="Volume horaire (heures)", y="Enseignement", hue="Type d'enseignement",
    ci=None, palette=["#d0e0e3", "#76a5af"], # palette="mako",
    ax=axs[0],
)
sns.despine(left=True, ax=axs[0])
axs[0].legend(bbox_to_anchor=(0.25, 1.2), loc='upper left', ncol=2)
axs[0].set_ylabel("")
axs[0].set_xlim([0, 60])
axs[0].tick_params(axis=u'y', which=u'both',length=0)
for c_idx, (index, row) in zip([-0.05, 0.35, 1.35, 2.35, 2.95, 3.35], df.iterrows()):
    g.text(
        row["Volume horaire (heures)"],
        c_idx,
        f'{row["Volume horaire (heures)"]}h ',
        ha="right",
    )

df_2 = df.groupby("Niveau").sum().reset_index()
g = sns.barplot(
    data=df_2,
    y="Niveau", x="Volume horaire (heures)",
    ci=None, color="#45818e", ax=axs[1]
)
sns.despine(left=True, ax=axs[1])
axs[1].set_ylabel("")
axs[1].set_xlim([0, 175])
axs[1].tick_params(axis=u'y', which=u'both',length=0)
for c_idx, (index, row) in zip([0.05, 1.05], df_2.iterrows()):
    g.text(
        row["Volume horaire (heures)"],
        c_idx,
        f'{row["Volume horaire (heures)"]}h ',
        ha="right",
    )

fig.subplots_adjust(wspace=0.3)
fig.suptitle("Enseignements realisés entre 2012 et 2018", y=1.2)
plt.savefig("enseignements.png", bbox_inches='tight')
```

```python
df = pd.DataFrame(
    {
        "Nombre de publications": [
            4, 12, 1
        ],
        "Type de publication": [
            "Revues internationales",
            "Conférence internationales",
            "Thèse",
        ],
    }
)
```

```python
df
```

```python
g = sns.barplot(
    data=df,
    y="Type de publication", x="Nombre de publications",
    ci=None, color="#45818e",
)
sns.despine(left=True)
plt.ylabel("")
plt.xlim([0, 15])
plt.tick_params(axis=u'y', which=u'both',length=0)
for c_idx, (index, row) in zip([0.1, 1.1, 2.1], df.iterrows()):
    g.text(
        row["Nombre de publications"],
        c_idx,
        f' {row["Nombre de publications"]}',
        ha="left",
    )
plt.title("Résumé des contributions scientifiques", y=1.1)
plt.savefig("publications.png", bbox_inches='tight')
```
